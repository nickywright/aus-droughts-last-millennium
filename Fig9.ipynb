### Dalculate drought relative intensity and severity

```python
def mean_intensity_severity_percentage(ds, threshold_start, threshold_end):
    ds_clim = ds.where((ds['year'] >= threshold_start) & (ds['year'] <= threshold_end), drop=True)
    ds['droughts_2s2e_intensity_pc'] = ds.droughts_2s2e_intensity / ds_clim.PRECT_mm.mean(dim='year')
    ds['droughts_2s2e_intensity_pc_mean'] = ds.droughts_2s2e_intensity_pc.mean(dim='year') * -100
    ds['droughts_2s2e_severity_pc'] = ds.droughts_2s2e_severity / ds_clim.PRECT_mm.mean(dim='year')
    ds['droughts_2s2e_severity_pc_mean'] = ds.droughts_2s2e_severity_pc.mean(dim='year') * -100
```

```python
mean_intensity_severity_percentage(hist_ens, 1900, 2000)
mean_intensity_severity_percentage(lm_ens, 1900, 2000)
```

### Calculate all ensemble means and differences between the two time periods

```python
hist_ens["droughts_2s2e_mean"] = xr.where(hist_ens.droughts_2s2e_mean >= 0.000000001, hist_ens.droughts_2s2e_mean, np.NaN)
hist_ens["droughts_2s2e_max"] = xr.where(hist_ens.droughts_2s2e_max >= 0.000000001, hist_ens.droughts_2s2e_max, np.NaN)

lm_ens["droughts_2s2e_mean"] = xr.where(lm_ens.droughts_2s2e_mean >= 0.000000001, lm_ens.droughts_2s2e_mean, np.NaN)
lm_ens["droughts_2s2e_max"] = xr.where(lm_ens.droughts_2s2e_max >= 0.000000001, lm_ens.droughts_2s2e_max, np.NaN)

hist_ens["droughts_2s2e_no_events_100yrs"] = xr.where(hist_ens.droughts_2s2e_no_of_events >= 0.000000001, hist_ens.droughts_2s2e_no_of_events, np.NaN)

lm_ens["droughts_2s2e_no_events_100yrs"] = lm_ens.droughts_2s2e_no_of_events/11.39
lm_ens["droughts_2s2e_no_events_100yrs"] = xr.where(lm_ens.droughts_2s2e_no_events_100yrs >= 0.000000001, lm_ens.droughts_2s2e_no_events_100yrs, np.NaN)

hist_ens['droughts_2s2e_sum_100yrs'] = hist_ens.drought_years_2s2e.sum(dim = 'year')
hist_ens['droughts_2s2e_sum_100yrs'] = xr.where(hist_ens.droughts_2s2e_sum_100yrs >= 0.000000001, hist_ens.droughts_2s2e_sum_100yrs, np.NaN)

lm_ens['droughts_2s2e_sum_100yrs'] = lm_ens.drought_years_2s2e.sum(dim = 'year')/11.39
lm_ens['droughts_2s2e_sum_100yrs'] = xr.where(lm_ens.droughts_2s2e_sum_100yrs >= 0.000000001, lm_ens.droughts_2s2e_sum_100yrs, np.NaN)
```

```python
# mean
lm_mean = lm_ens['droughts_2s2e_mean'].mean(dim='model')
hist_mean = hist_ens['droughts_2s2e_mean'].mean(dim='model')
diff_mean = hist_mean - lm_mean

# max
lm_max = lm_ens['droughts_2s2e_max'].mean(dim='model')
hist_max = hist_ens['droughts_2s2e_max'].mean(dim='model')
diff_max = hist_max - lm_max

# intensity
lm_int = lm_ens['droughts_2s2e_intensity_pc_mean'].mean(dim='model')
hist_int = hist_ens['droughts_2s2e_intensity_pc_mean'].mean(dim='model')
diff_int = hist_int - lm_int

# severity
lm_sev = lm_ens['droughts_2s2e_severity_pc_mean'].mean(dim='model')
hist_sev = hist_ens['droughts_2s2e_severity_pc_mean'].mean(dim='model')
diff_sev = hist_sev - lm_sev

# frequency
lm_freq = lm_ens['droughts_2s2e_no_events_100yrs'].mean(dim='model')
hist_freq = hist_ens['droughts_2s2e_no_events_100yrs'].mean(dim='model')
diff_freq = hist_freq - lm_freq

# proportion of time in drought
lm_prop = lm_ens['droughts_2s2e_sum_100yrs'].mean(dim='model')
hist_prop = hist_ens['droughts_2s2e_sum_100yrs'].mean(dim='model')
diff_prop = hist_prop - lm_prop
```

## Perform t-tests: are pre-industrial and 20th century different?

```python
# details here! https://github.com/scipy/scipy/blob/v0.19.0/scipy/stats/stats.py#L3769-L3846

# with a little bit of help from https://gist.github.com/kuchaale/293d2a16726a5d492be4f5bbae8d9111
```

```python
lat_ref = lm_ens['droughts_2s2e_mean'].lat
lon_ref = lm_ens['droughts_2s2e_mean'].lon

# mean
tstat, pval_mean = ttest_rel(lm_ens['droughts_2s2e_mean'], hist_ens['droughts_2s2e_mean'])
sig_mean = np.where(pval_mean < 0.05, 1, 0)
sig_mean_da = xr.DataArray(sig_mean, coords=[lat_ref, lon_ref], dims=['lat', 'lon'])

# max
tstat, pval_max = ttest_rel(lm_ens['droughts_2s2e_max'], hist_ens['droughts_2s2e_max'])
sig_max = np.where(pval_max < 0.05, 1, 0)
sig_max_da = xr.DataArray(sig_max, coords=[lat_ref, lon_ref], dims=['lat', 'lon'])

# intensity
tstat, pval_int = ttest_rel(lm_ens['droughts_2s2e_intensity_pc_mean'], hist_ens['droughts_2s2e_intensity_pc_mean'])
sig_int = np.where(pval_int < 0.05, 1, 0)
sig_int_da = xr.DataArray(sig_int, coords=[lat_ref, lon_ref], dims=['lat', 'lon'])

# severity
tstat, pval_sev = ttest_rel(lm_ens['droughts_2s2e_severity_pc_mean'], hist_ens['droughts_2s2e_severity_pc_mean'])
sig_sev = np.where(pval_sev < 0.05, 1, 0)
sig_sev_da = xr.DataArray(sig_sev, coords=[lat_ref, lon_ref], dims=['lat', 'lon'])

# freq
tstat, pval_freq = ttest_rel(lm_ens['droughts_2s2e_no_events_100yrs'], hist_ens['droughts_2s2e_no_events_100yrs'])
sig_freq = np.where(pval_freq < 0.05, 1, 0)
sig_freq_da = xr.DataArray(sig_freq, coords=[lat_ref, lon_ref], dims=['lat', 'lon'])

# prop
tstat, pval_prop = ttest_rel(lm_ens['droughts_2s2e_sum_100yrs'], hist_ens['droughts_2s2e_sum_100yrs'])
sig_prop = np.where(pval_prop < 0.05, 1, 0)
sig_prop_da = xr.DataArray(sig_prop, coords=[lat_ref, lon_ref], dims=['lat', 'lon'])
```
